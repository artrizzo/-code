```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
Grafo_1 = nx.Graph()
Grafo_1.add_nodes_from(["a","b","c",'d',"e","f","g","h","k","m","n","p","q","r"])
```

```python
Grafo_1.add_edges_from([("m","n"), ("n","f"), ("f","b"), ("b", "c"), ("c", "h"), ("h", "g"), ("h", "k"), ("g", "p"), ("g", "q"), ("p", "q"), ("q", "r"), ("q", "k"), ("r", "k"), ("k", "k"), ("d", "e")])
```

```python
nx.draw(Grafo_1, with_labels=True, node_size=200, node_color='gold', font_weight='bold')
```

```python
#a - numero de componentes conexos
#b - ordem e tamanho da maior componente conexa
#c - ordem e tamanho da menor componente conexa
#d - grau dos vértices a,k,q
#e - utilize o algoritmo de busca de largura para obter a distância de a até q e de c até q
#f - como verificar se o grafo é não conexo?
#g - como saber quais vértices são alcançáveis a partir de c?
```

```python
#a
print(Grafo_1.degree())
```

```python
#b
print("Ordem= ", Grafo_1.order())
print("Tamanho = ", Grafo_1.size())
```

```python
#c
```

```python
#d
print(Grafo_1.degree("a"))
print(Grafo_1.degree("k"))
print(Grafo_1.degree("q"))
```

```python
#e
for u, v in Grafo_1.edges:
  print(u, v)

nx.shortest_path(Grafo_1, 'a', 'q')
nx.shortest_path(Grafo_1, 'c', 'q')
```

```python
#f
nx.is_connected(Grafo_1)
```

```python
#g
nx.shortest_path(Grafo_1, 'c')
```
