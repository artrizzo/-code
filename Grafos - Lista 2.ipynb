```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
nx.__version__
```

```python
#Criando um grafo vazío
G = nx.Graph()
```

```python
# retorna uma lista de vértices
print(G.nodes)
```

```python
# retorna uma lista de arestas
print(G.edges)
```

```python
G.add_node("A")
```

```python
G.add_nodes_from(["B","C",'D',"E"])
```

```python
print(G.nodes)
```

```python
G.add_edge("A","B")
```

```python
G.add_edges_from([("A","C"), ("B","D"), ("B","E"), ("C", "E")])
```

```python
print(G.edges)
```

```python
nx.draw(G)
```

```python
nx.draw(G)
plt.savefig("grafo.png")
```

```python
#insere as letras nos círculos azuis
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
#Está ligado a quem (adjacente a quem)
print(G.adj)
```

```python
#grau
print(G.degree("A"))
```

```python
#grau de todos
print(G.degree())
```

```python
G = nx.Graph() 

E = [('A', 'B', 2), ('A', 'C', 1), ('B', 'D', 5), ('B', 'E', 3), ('C', 'E', 2)]

G.add_weighted_edges_from(E)
```

```python
pos = nx.spring_layout(G);

nx.draw(G, pos, with_labels=True, font_weight='bold');

edge_weight = nx.get_edge_attributes(G,'weight');

nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_weight);

#plt.savefig("grafoPonderado.png")
```

```python
#cógigos da documentação oficial
G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2, 3])
print(G.nodes)
```

```python
G.add_nodes_from([
   (4, {"color": "red"}),
   (5, {"color": "green"}),
])
print(G.nodes)
```

```python
H = nx.path_graph(10)
G.add_nodes_from(H)
print(H.nodes)
print(H.edges)
```

```python
nx.draw(H)
```

```python
#adicionando arestas
G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)
G.add_edges_from([(1, 3), (1, 4)])
print(G.edges)
```

```python
G.add_edges_from(H.edges)
print(G.edges)
nx.draw(G)
```

```python
#removendo todas as arestas de um grafo
G.clear_edges()
print(G.nodes)
print(G.edges)
#ou
G.clear()
```
