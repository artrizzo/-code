```python
#Escreva um programa que calcule a idade de uma pessoa a partir do seu ano de nascimento.
#O programa deve mostrar na tela a idade e tamb√©m as seguintes mensagens, quando for o caso:
#"Ja pode votar" (se tiver 16 anos ou mais)
#"Ja pode solicitar a Carteira de Habilitacao" (se tiver 18 anos ou mais)

nascimento = int(input('escreva o ano de nascimento: '))
currentDateTime = datetime.datetime.now()
date = currentDateTime.date()
ano_atual = int(date.strftime("%Y"))
idade = ano_atual - nascimento
print(idade) 

if idade>=16:
  print("Ja pode votar")
elif idade>=18:
    print("Ja pode solicitar a Carteira de Habilitacao")
else:
  print("N√£o pode votar")
```

```python
#Escreva um algoritmo que pe√ßa como entrada os coeficientes a, b e c (n√∫meros reais) de uma equa√ß√£o de 2.o grau.
#Suponha que o usu√°rio sempre digita um n√∫mero positivo para o coeficiente a. O programa deve fornecer como sa√≠da:
#I. se Œî = 0 : Calcule e imprime a √∫nica raiz pela f√≥rmula de Bhaskara
#II. se Œî < 0 : Imprima a mensagem: "As duas raizes sao complexas"
#III. se Œî > 0 : Calcule e imprime as 2 ra√≠zes reais (com duas casas decimais) pela f√≥rmula de Bhaskara (Œî=ùëè2‚àí4ùëéùëê)

a = int(input('escreva um n√∫mero: '))
b = int(input('escreva um n√∫mero: '))
c = int(input('escreva um n√∫mero: '))

delta = (b**2 - 4*a*c)

raiz_1 = (-b + (delta**1/2)/2*a)

raiz_2 = (-b - (delta**1/2)/2*a)

if delta == 0:
  print((-b+((delta**1/2)))/2*a)

if delta > 0:
  print('As duas raizes s√£o complexas')

if delta < 0:
  print('\nAs duas raizes s√£o %2.f e %2.f'%(raiz_1, raiz_2))
```
