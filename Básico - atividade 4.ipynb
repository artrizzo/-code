```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
prova = pd.read_csv('https://drive.google.com/u/3/uc?id=19KUegmeTVO8cWLvH7_2eB322o8uWH8MC&export=download', sep=',')
```

```python
# Leia a planilha e calcule o coeficiente de correlação entre:
# Prova 1 e Prova 2 (2 casas decimais)
corr1 = prova['Prova 1'].corr(prova['Prova 2'])  
print(f'{corr1:.2}')
```

```python
# Prova 1 e Trabalho (2 casas decimais)
corr2 = prova['Prova 1'].corr(prova['Trabalho'])
print(f'{corr2:.2}')
```

```python
# Prova 2 e Trabalho (2 casas decimais)
corr3 = prova['Prova 2'].corr(prova['Trabalho'])
print(f'{corr3:.2}')
```

```python
# Qual é o par mais fortemente correlacionado?
print(f'O par mais fortemente correlacionado é: {max(corr1, corr2, corr3):.2}')
```

```python
# Ler a planilha e uma das colunas "Prova 1" ou "Prova 2",
# calcular e imprimir o coeficiente de correlação da coluna Trabalho com a
# coluna lida. 2 casas decimais.
coluna = input()
corr_coluna = prova['Trabalho'].corr(prova[coluna])
corr_coluna
```

```python
#O que podemos concluir a partir deste valor obtido?
print(f'A correlação entre Trabalho e {coluna} é {corr_coluna:.2f}')
```

```python
# Leia a planilha
# Quais são os valores de A e B para a reta A*x+B de regressão linear com a Prova 1 no eixo x e a Prova 2 no eixo y?
(a, b) = np.polyfit(x=prova["Prova 1"], y=prova["Prova 2"], deg = 1)
print(a)
print(b)
```

```python
# Imprime a reta de regressão no formato y=A*x+B, com A e B os valores obtidos no passo acima,agora com 3 casas decimais
# (note que não há espaço em branco entre A e x nesta formatação).
x = np.arange(0,51)
y = a*x+b
plt.plot(x,y,'r')
```

```python
# Qual seria a predição baseada nessa reta para a nota na Prova 2 se a nota na Prova 1 for 65?
x = np.arange(0,65)
y = a*x+b
plt.plot(x,y,'r')
```
