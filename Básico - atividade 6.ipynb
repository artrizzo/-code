```python
#Faça um programa que pede para o usuário digitar uma senha. Ele deve ter no máximo 3 tentativas. A senha correta é 1234 Em cada tentativa,
# a - Se ele acertar, diga que ele acertou e encerre o programa.
# b - Se ele errar, diga que a senha está errada e dê mais uma tentativa (com um limite de 3 tentativas)

chances = 1

while chances <= 3:
  senha = str(input('Digite uma senha: '))
  if senha == '1234':
    print('Você acertou a senha.')
    break
  else:
    print('Tente novamente')
  chances = chances + 1 
```
